```python
import sklearn as ski
from skimage import data, io, filters
import numpy as np
import matplotlib.pyplot as plt

plt.set_cmap("gray")

img = data.camera()
blured = filters.gaussian(img, sigma=10)

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(10, 5))
ax[0].imshow(img)
ax[1].imshow(blured)
ax[2].imshow(filters.unsharp_mask(blured, radius=3, amount=8))
```

```python
import sklearn as ski
from skimage import data, io, filters
import numpy as np
import matplotlib.pyplot as plt

plt.set_cmap("gray")

img = data.text()

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
ax[0].imshow(img)
ax[1].imshow(filters.sobel(img))
```

```python
import sklearn as ski
from skimage import data, io, filters
from skimage.exposure import histogram as hist
import numpy as np
import matplotlib.pyplot as plt

plt.set_cmap("gray")

img = data.camera()
treshold = filters.threshold_otsu(img)

fig, ax = plt.subplots(1,3,figsize=(15,5))
ax[0].imshow(img)
ax[1].plot(hist(img)[0])
ax[2].imshow(img>treshold)
```
